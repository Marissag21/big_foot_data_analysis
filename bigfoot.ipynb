```python
# Load in dependencie
import pandas as pd
import numpy as np
from datetime import datetime as dt
```

```python
# Load in file
bigfoot_csv_path = "Resources/bfro_reports_geocoded.csv"
```

```python
# Read and display the CSV with Pandas
bigfoot_file_df = pd.read_csv(bigfoot_csv_path, header=0)
bigfoot_file_df.head()
```

```python
# List all the columns in the table
bigfoot_file_df.columns
```

```python
# Create a new table that takes clean data
df2=bigfoot_file_df.dropna(subset=['number','county','state','latitude','longitude','date','season','classification',\
    'temperature_mid','humidity','cloud_cover','precip_intensity','visibility','pressure','wind_speed','observed'])
clean_bigfoot_df=df2.dropna().reset_index(drop=True)
clean_bigfoot_df.head()

```

```python
#Using drop() to delete unknown rows in season column 
clean_bigfoot_df.drop(clean_bigfoot_df[clean_bigfoot_df['season'] == 'Unknown'].index, inplace = True)
clean_bigfoot_df
```

```python
# Columns we're interested in:
final_df = clean_bigfoot_df[['number','county','state','latitude','longitude','classification','date','season',\
    'temperature_mid','humidity','cloud_cover','precip_intensity','visibility','pressure','wind_speed','observed']]
final_df.head()
```

```python
#Take a look to see what types each column is
final_df.dtypes
```

```python
#Take a look and make sure each column has the same amount of rows
final_df.count()
```

```python
#Rename columns so they are easier to work with
final_df = final_df.rename(
    columns={'number':'Number','county':'County','state':'State','latitude':'Latitude',\
        'longitude':'Longitude','classification':'Classification','date':'Date','season':'Season',\
        'temperature_mid':'Temperature','humidity':'Humidity','cloud_cover':'Cloud_cover','precip_intensity':\
            'Precip_intensity','visibility':'Visibility','pressure':'Pressure','wind_speed':'Wind_speed','observed':'Observed'})
final_df.head()
```

```python
# Change number & temperature_mid to int
# Change the rest of the columns to strings
final_df['Number'] = final_df['Number'].astype('int64')
final_df["Temperature"] = final_df['Temperature'].astype('int64')
final_df['County'] = final_df['County'].astype('string')
final_df['State'] = final_df['State'].astype('string')
final_df['Classification'] = final_df['Classification'].astype('string')
final_df['Date'] = final_df['Date'].astype('string')
final_df['Season'] = final_df['Season'].astype('string')
final_df['Observed'] = final_df['Observed'].astype('string')

final_df.head()
```

```python
# Making sure all the data types have been changed
final_df.dtypes
```

```python
# Export Pandas DataFrame to CSV
final_df.to_csv('bigfoot_clean_file.csv', encoding='utf-8',index=False)
```
